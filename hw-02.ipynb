Import necessary libraries

```python
from matplotlib import pyplot as plt
import numpy as np
```

We first define the density function

```python
def f(x, a=0.2):
    return (1/a) * x**(-(1+a)/a)
```

Plot for various values of a

```python
x = np.linspace(0.01, 0.1, 1000)
for a in [0.5, 0.55, 0.6]:
    plt.plot(x, f(x, a), label='a={}'.format(a))
    plt.legend()
    plt.grid()
plt.show()
```

```python
x = np.linspace(0.01, 0.5, 1000)
for a in [100, 200, 300, 400, 500]:
    plt.plot(x, f(x, a), label='a={}'.format(a))
    plt.legend()
    plt.grid()
plt.show()
```
