Import necessary libraries

```python
from matplotlib import pyplot as plt
import numpy as np
```

We first define the density function

```python
def f(x, a):
    return (1/a) * x**(-(1+a)/a)
```

Plot for various values of a

```python
x = np.linspace(1, 8, 1000)
for a in [0.2, 0.5, 1, 2, 5]:
    plt.plot(x, f(x, a), label='a={}'.format(a))
    plt.legend()
    plt.grid()
plt.show()
```

Notice how fast the function is dropping when α is a small value. For larger values of α, the decrease is not so rapid. The comparison does make sense intuitively.
